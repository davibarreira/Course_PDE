# Questão 2

## a)

```julia
Δh = 0.1
n = 11
y = zeros(n)
t = zeros(n) 
f(t,y) = y+5
for i in 1:n-1
    y[i+1] = y[i] + Δh * f(t[i],y[i])
end
yreal(t) = 5*exp(t) - 5
```

**Resultado numérico**

```julia
y
```

**Resultado real**

```julia
yreal.(0:0.1:1)
```

# Questão 3

```julia
using CairoMakie
```

## a)

```julia
xt(x,y) = 1
yt(x,y) = 2*y + 4 - x
odeSol(x,y) = Point(xt(x,y),yt(x,y)) # x'(t) = -x, y'(t) = 2y
min = -5
max = 5
xs = LinRange(min, max, 20)
ys = LinRange(min, max, 10)
us = [xt(x,y) for x in xs, y in ys]
vs = [yt(x,y) for x in xs, y in ys]

fig = Figure(resolution = (500, 350), fontsize = 18, font = "sans")
ax = Axis(fig, xlabel = "x", ylabel = "y")
arrows!(ax, xs, ys, us, vs, arrowsize = 8, lengthscale = 0.1, color=:green)
streamplot!(ax, odeSol, min..max, min..max,gridsize= (30,30),
    arrow_size = 0, linewidth=2, colormap = (:black, :black))
limits!(ax, min,max,min,max)
fig[1, 1] = ax
save("3a.png", fig)
fig
```
